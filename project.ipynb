as seen in DEA the variable **ca** (Number of major vessels (0–3) colored by flourosopy) has **4 missing values** and the variable **thal** (thalassemia) has **2 missing values** 

scince both variables (columns) are contains categorical data we will replace the missing values with the **mode** of each variable

```python
df.num = df.num.astype(bool)
```

# Data Transformation

--------------------------------------------------------------------


### Data Train Test Split

```python
categorical = ['sex','cp','thal','ca','fbs','restecg','exang','slope']
X_categorical = df[categorical].values

x = df.drop(['num'], axis = 1).values
y = df.num.values
X = df.drop(categorical+['num'], axis = 1).values
```

### Data Transformation

```python
corr = df_corr['num'].drop(categorical+['num']).abs().values
```

```python
# Convert Categorical data to dummy data 
ohe = OneHotEncoder(sparse=False, drop='first')
X_ohe = ohe.fit_transform(X_categorical)
```

```python
# Standarization for the numirical data (z-score)
X_standrized = StandardScaler().fit_transform(X)
```

```python
# Add weights to the numical data (from corrolation matrix)
X_standrized_C = X_standrized * df_corr['num'].drop(categorical+['num']).abs().values
```

```python
# combine the numirical data with the categorical data 
X_standrized_C = np.concatenate((X_standrized_C,X_ohe),axis=1)
X = np.concatenate((X,X_ohe),axis=1)
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split ( x, y,test_size = 0.3, random_state=101)

XC_train, XC_test, yc_train, yc_test = train_test_split ( X_standrized_C, y,test_size = 0.3, random_state=101)
```

<a id="ref8"></a>
# Data Modeling

-------------------------------------------------------------------------------

### Logistic Regression

```python
logmodel = LogisticRegression()
logmodel.fit(XC_train,yc_train)
prediction_logmodel = logmodel.predict(XC_test)

print(classification_report(y_test,prediction_logmodel))
lrm = (confusion_matrix(yc_test,prediction_logmodel))
```

```python
# Cross validation n=5
cvscore_log=cross_val_score(logmodel, X_standrized_C, y, cv=5, n_jobs=-1)
accuracy_test_logmodel = round(cvscore_log.mean(),3)
accuracy_test_logmodel 
```

### KNN Model

```python
import sklearn.metrics as metrics
# Finding the best K
Ks = 25
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))
ConfustionMx = [];
for n in range(1,Ks):
    
    #Train Model and Predict  
    neigh = KNeighborsClassifier(n_neighbors = n).fit(XC_train,yc_train)
    yhat=neigh.predict(XC_test)
    mean_acc[n-1] = metrics.accuracy_score(yc_test, yhat)

    
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])

mean_acc.max()
```

```python
k_best = mean_acc.argmax()+1
knnmodel = KNeighborsClassifier(n_neighbors=k_best)
knnmodel.fit(X_train,y_train)
prediction_knnmodel = knnmodel.predict(X_test)

print(classification_report(y_test,prediction_knnmodel))
knnm = (confusion_matrix(y_test,prediction_knnmodel))
```

```python
# Cross validation n=5
cvscore_knn=cross_val_score(knnmodel, X_standrized_C, y, cv=5, n_jobs=-1)
accuracy_test_knnmodel = round(cvscore_knn.mean(),3)
accuracy_test_knnmodel
```

### Decision Tree Model 

```python
dtreemodel = DecisionTreeClassifier(criterion="entropy", max_depth = 2)
dtreemodel.fit(X_train,y_train)
prediction_dtreemodel = dtreemodel.predict(X_test)

# accuracy_test_dtreemodel = round(dtreemodel.score(X_test, y_test) * 100, 2)

print(classification_report(y_test,prediction_dtreemodel))
dtm = (confusion_matrix(y_test,prediction_dtreemodel))
```

```python
# Cross validation n=5
cvscore_dtree=cross_val_score(dtreemodel, X, y, cv=5, n_jobs=-1)
accuracy_test_dtreemodel = round(cvscore_dtree.mean(),3)
accuracy_test_dtreemodel
```

### Random Forest Model

```python
rfcmodel = RandomForestClassifier(n_estimators=600)
rfcmodel.fit(X_train,y_train)
prediction_rfcmodel = rfcmodel.predict(X_test)

# accuracy_test_rfcmodel = round(rfcmodel.score(X_test, y_test) * 100, 2)

print(classification_report(y_test,prediction_rfcmodel))
rfm = (confusion_matrix(y_test,prediction_rfcmodel))
```

```python
# Cross validation n=5
cvscore_rfc=cross_val_score(rfcmodel, X, y, cv=5, n_jobs=-1)
accuracy_test_rfcmodel = round(cvscore_rfc.mean(),3)
accuracy_test_rfcmodel
```

### Support Vector Machine

```python
svcmodel = SVC(kernel='rbf', probability=True)
svcmodel.fit(XC_train,yc_train)
prediction_svcmodel = svcmodel.predict(XC_test)


print(classification_report(yc_test,prediction_svcmodel))
svmm = (confusion_matrix(yc_test,prediction_svcmodel))
```

```python
# Cross validation n=5
cvscore_svc=cross_val_score(svcmodel, X_standrized_C, y, cv=5, n_jobs=-1)
accuracy_test_svcmodel = round(cvscore_svc.mean(),3)
accuracy_test_svcmodel
```

### Naive Bayes

```python
nbmodel = MultinomialNB() 
nbmodel.fit(X_train, y_train)  
prediction_nbmodel = nbmodel.predict(X_test)

print(classification_report(y_test,prediction_nbmodel))
nbm = (confusion_matrix(y_test,prediction_nbmodel))
```

```python
# Cross validation n=5
from sklearn.model_selection import cross_val_score
cvscore_nb=cross_val_score(nbmodel, X, y, cv=5, n_jobs=-1)
accuracy_test_nbmodel = round(cvscore_nb.mean(),3)
accuracy_test_nbmodel
```

### Ridge Classifier

```python
ridgemodel = RidgeClassifier()
ridgemodel.fit(X_train, y_train)
prediction_ridgemodel = ridgemodel.predict(X_test)

rcm = confusion_matrix(y_test,prediction_ridgemodel)
print(classification_report(y_test,prediction_ridgemodel))
```

```python
# Cross validation n=5
cvscore_ridge=cross_val_score(ridgemodel, X, y, cv=5, n_jobs=-1)
accuracy_test_ridgemodel = round(cvscore_ridge.mean(),3)
accuracy_test_ridgemodel
```
